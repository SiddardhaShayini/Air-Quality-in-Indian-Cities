#### Let's look at the missing values

```python
# Missing values
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()

        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(df)

        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)

        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})

        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)

        # Print some summary information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")

        # Return the dataframe with missing information
        return mis_val_table_ren_columns

missing_values= missing_values_table(df)
missing_values.style.background_gradient(cmap='Reds')
```

```python
import pandas as pd
from IPython.display import display, HTML
```

```python
def max_polluted_city(pollutant):
    """
    Calculates and returns the top 10 most polluted cities for a given pollutant,
    styled with a background gradient.
    """
    x1 = city_day[[pollutant,'City']].groupby(["City"]).mean().sort_values(by=pollutant,ascending=False).reset_index()
    x1[pollutant] = round(x1[pollutant],2)
    return x1[:10].style.background_gradient(cmap='OrRd')
```

```python
def display_side_by_side(*args):
    """
    Displays multiple Styler objects (or DataFrames) side-by-side using HTML.
    """
    html_str = ''
    for df_or_styler in args:
        # Check if the object has the _repr_html_ method (which Styler objects do)
        if hasattr(df_or_styler, '_repr_html_'):
            html_str += df_or_styler._repr_html_()
        elif isinstance(df_or_styler, pd.DataFrame):
            # If a DataFrame is passed, convert it to Styler and then get HTML
            html_str += df_or_styler.style._repr_html_()
        else:
            # Fallback for other types, convert to string
            html_str += str(df_or_styler)

    # Use inline-block display for better side-by-side arrangement
    # and add some margin for separation
    final_html = html_str.replace('<table', '<div style="display:inline-block; margin: 10px;"><table')
    final_html = final_html.replace('</table>', '</table></div>')

    display(HTML(final_html))
```

```python
# Generate the styled DataFrames
pm2_5 = max_polluted_city('PM2.5')
pm10 = max_polluted_city('PM10')
no2 = max_polluted_city('NO2')
so2 = max_polluted_city('SO2')
co = max_polluted_city('CO')
btx = max_polluted_city('BTX')

# Display them side-by-side
display_side_by_side(pm2_5, pm10, no2, so2, co, btx)
```

```python
# Copy original DataFrame
df1 = df.copy()

# Fill missing values with median for numeric columns
numeric_cols = [
    'PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3',
    'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI'
]

for col in numeric_cols:
    df1[col] = df1[col].fillna(df1[col].median())

# Fill missing for AQI_Bucket
df1['AQI_Bucket'] = df1['AQI_Bucket'].fillna('Moderate')
# Add total pollution metric
df1['Pollution Content'] = df1[numeric_cols[:-1]].sum(axis=1)  # exclude AQI from sum
```

```python
# Median pollution by city (line plot)
def plot_pollution_line():
    city_pollution = df1.groupby('City')['Pollution Content'].median().sort_values(ascending=False)
    city_pollution.plot(marker='o', color='black')
    plt.title('Median Pollution Content Across Cities')
    plt.xlabel('City')
    plt.ylabel('Pollution Content')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
# Top 10 polluted cities by median
def max_pollution_bar():
    top_cities = df1.groupby('City')['Pollution Content'].median().sort_values(ascending=False).head(10)

    # Debug check
    print("\nTop Polluted Cities Data:\n", top_cities)

    if top_cities.empty:
        print("No data to plot for top polluted cities.")
        return

    sns.barplot(x=top_cities.index, y=top_cities.values, palette='Reds')
    plt.title('Top 10 Most Polluted Cities')
    plt.xlabel('City')
    plt.ylabel('Median Pollution Content')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Least polluted cities by mean
def min_pollution_bar():
    bottom_cities = df1.groupby('City')['Pollution Content'].mean().sort_values(ascending=True).head(10)

    # Debug check
    print("\nLeast Polluted Cities Data:\n", bottom_cities)

    if bottom_cities.empty:
        print("No data to plot for least polluted cities.")
        return

    sns.barplot(x=bottom_cities.index, y=bottom_cities.values, palette='Greens')
    plt.title('Top 10 Least Polluted Cities')
    plt.xlabel('City')
    plt.ylabel('Mean Pollution Content')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

### Modeling

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/SiddardhaShayini/Air-Quality-in-Indian-Cities/refs/heads/main/dataset/city_day.csv')
df
```

```python
# Ensure Date is datetime
df['Date'] = pd.to_datetime(df['Date'])

# Fill missing numeric values with median
pollutants = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene']
for col in pollutants + ['AQI']:
    df[col] = df[col].fillna(df[col].median())

df['Total Pollution'] = df[pollutants].sum(axis=1)

# Extract year and month
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
```

```python
# Feature and target
features = pollutants + ['Total Pollution', 'Year', 'Month']
X = df[features]
y = df['AQI']
```

```python
# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Train model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np


y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae:.4f}")

mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse:.4f}")

rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")

r2 = r2_score(y_test, y_pred)
print(f"R-squared (R2 Score): {r2:.4f}")

```

```python
def predict_city_year(city_name, year):
    city_data = df[df['City'] == city_name]
    monthly_avg = city_data.groupby(['Year', 'Month'])[pollutants].mean().reset_index()
    monthly_avg['Total Pollution'] = monthly_avg[pollutants].sum(axis=1)
    monthly_avg['Year'] = year  # simulate future year

    X_pred = monthly_avg[features]
    pred_aqi = model.predict(X_pred)
    monthly_avg['Predicted AQI'] = pred_aqi
    monthly_avg['AQI Bucket'] = monthly_avg['Predicted AQI'].apply(get_aqi_bucket)

    print(f"\n📍 Predicted AQI for {city_name} in {year}:")
    print(monthly_avg[['Month', 'Predicted AQI', 'AQI Bucket']])

    # Plot
    plt.figure(figsize=(10, 4))
    plt.plot(monthly_avg['Month'], monthly_avg['Predicted AQI'], marker='o', color='orange')
    plt.title(f"{city_name} - Predicted Monthly AQI for {year}")
    plt.xlabel('Month')
    plt.ylabel('AQI')
    plt.grid(True)
    plt.xticks(range(1,13))
    plt.tight_layout()
    plt.show()
```

```python
def top10_cities_by_year(year):
    top_data = []

    for city in df['City'].unique():
        city_data = df[df['City'] == city]
        monthly_avg = city_data.groupby(['Year', 'Month'])[pollutants].mean().reset_index()
        monthly_avg['Total Pollution'] = monthly_avg[pollutants].sum(axis=1)
        monthly_avg['Year'] = year

        if not monthly_avg.empty:
            X_pred = monthly_avg[features]
            pred_aqi = model.predict(X_pred)
            avg_aqi = np.mean(pred_aqi)
            top_data.append((city, avg_aqi))

    top_df = pd.DataFrame(top_data, columns=['City', 'Predicted AQI'])
    top_df['AQI Bucket'] = top_df['Predicted AQI'].apply(get_aqi_bucket)
    top_df = top_df.sort_values(by='Predicted AQI', ascending=False).head(10)

    print(f"\n🔥 Top 10 Most Polluted Cities in {year} (Predicted AQI):")
    print(top_df)

    # Plot
    plt.figure(figsize=(10, 5))
    plt.bar(top_df['City'], top_df['Predicted AQI'], color='red')
    plt.title(f"Top 10 Polluted Cities in {year}")
    plt.ylabel("Predicted AQI")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
def get_aqi_bucket(aqi):
    if aqi <= 50: return 'Good'
    elif aqi <= 100: return 'Satisfactory'
    elif aqi <= 200: return 'Moderate'
    elif aqi <= 300: return 'Poor'
    elif aqi <= 400: return 'Very Poor'
    else: return 'Severe'
```

```python
def menu():
    print("\n📋 Select Option:")
    print("1. Predict AQI for a Specific City and Year")
    print("2. Show Top 10 Polluted Cities for a Year")
    print("3. Exit")

    choice = input("\nEnter your choice (1/2/3): ")

    if choice == '1':
        city = input("Enter city (e.g., Delhi): ").strip()
        year = int(input("Enter year (e.g., 2026): "))
        predict_city_year(city, year)

    elif choice == '2':
        year = int(input("Enter year (e.g., 2026): "))
        top10_cities_by_year(year)

    elif choice == '3':
        print("✅ Exiting.")

    else:
        print("❌ Invalid choice. Exiting.")

menu()
```
